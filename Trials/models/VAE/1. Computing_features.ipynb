# Wav파일을 Mel spectrogram으로 feature구해서 저장하는 파일
### input data 형식이 (row, 128, 48)
### wav -> mel spectrogrma -> (row, 128, 48) 형태로 np파일 저장

```python
%matplotlib inline
```

```python
import os

import numpy as np
import matplotlib.pyplot as plt

from scipy import signal
from scipy.io import wavfile
```

```python
# train test 일단 분리한 채로 하는것
train_dataset_path = '../son/train_5seconds/'
test_dataset_path = '../son/test_5seconds/'
train_features_path = '../son/feature_train/'
test_features_path = '../son/feature_test/'
```

```python
# 세부기능은 일단 차치하고 wav파일의 피쳐들을 구하는 것이다
import librosa

def generate_VAE_features(dataset_path, features_path):
    for folder in sorted(os.listdir(dataset_path)):
        for filename in os.listdir(os.path.join(dataset_path, folder)):
            # 일단 손석희 데이터로만 학습
            # 전체데이터로 할경우 if문 제거
            if folder == 'son':
                print(folder, filename)
                y, sr = librosa.load(os.path.join(dataset_path, folder, filename))
                # 한개 데이터로 할경우 주석해제
                # y, sr = librosa.load('../son/train_5seconds/son/NB10584578.wav_0.05.wav')
                S = librosa.feature.melspectrogram(y=y, sr=sr)
                print("Wav length: {}, Mel_S shape:{}".format(len(y)/sr, np.shape(S)))

                # split 지점 만들기: col 48의 배수인 지점마다
                # 학습시키기 위해 형식 맞춰주는 것임
                i=0
                _split = []
                while i < S.shape[1] - 48:
                    i += 48
                    _split.append(i)

                # split 기준에 따라 numpy 배열 분리 마지막거는 완벽한 모양이 아니라 빼주기 
                features = np.stack(np.hsplit(S, _split)[:-1])
                print(features.shape)
                # 각 wav파일의 feature 저장
                features.dump(os.path.join(features_path, filename.split('.')[0]))
            # mel spectrogram이 제대로 reverse 되는지 확인
#             last = librosa.feature.inverse.mel_to_audio(np.hstack(features))
#             librosa.output.write_wav('./mel__spec_output.wav', last, sr=22050)                    
```

```python
%%time
# train feature 만들기
generate_VAE_features(train_dataset_path, train_features_path)
```

```python
%%time
# test feature 만들기
generate_VAE_features(test_dataset_path, test_features_path)
```

```python
# 임시 파일 하나만 가지고 스펙토그램 그리고
# smapler rate -> framesize -> overlap size -> n_fft_points
FRAME_DURATION = 0.02

# sample_rate, samples = wavfile.read(os.path.join(train_dataset_path, '14-208-0000.wav'))
samples, sample_rate = librosa.load(os.path.join(train_dataset_path, 'son', 'NB10588435.wav_0.05.wav'))
frame_size = int(round(FRAME_DURATION * float(sample_rate)))
overlap_size = frame_size - int(round((FRAME_DURATION - 0.01) * float(sample_rate)))
print(frame_size)
n_fft_points = 2
while n_fft_points < frame_size:
    n_fft_points *= 2
print(n_fft_points)
frequencies, times, spectrogram = signal.spectrogram(
    samples, fs=sample_rate, window='hanning', nperseg=frame_size, noverlap=overlap_size, nfft=n_fft_points,
    scaling='spectrum', mode='psd'
)

plt.figure(figsize=(100, 20))
plt.pcolormesh(times, frequencies, 10*np.log10(spectrogram))
plt.savefig('../data/audio.png')
plt.show()
```
